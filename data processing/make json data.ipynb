```python
useful_nouns = useful[(useful["POS"]=="n") & (useful["ranking"]<3000)]
```

```python
useful_nouns.reset_index(inplace=True)
```

```python
useful_nouns.shape
```

```python
def make_set(start,end,set_name):
    json_extract = f'"{set_name}":{{\n'
    for index, row in useful_nouns[start:end].iterrows():
        print(index)

        new_entry = f'''\t"{row['English']}":"{row['Finnish']}"'''
        if index-start < end-start-1:
            new_entry+= ", \n"
        else:
            new_entry+="\n"
        json_extract += new_entry
    json_extract+="\t}, \n"
   
    return json_extract
```

```python
print(make_set(10,17,'bob'))
```

```python
with open("jsonnouns.txt", "w") as j:
    for i in range(0,2000,10):
        j.write(make_set(i,i+10,f"nouns{int(i/10)}"))
        print(i)
        print(i+10)
        
    
```
